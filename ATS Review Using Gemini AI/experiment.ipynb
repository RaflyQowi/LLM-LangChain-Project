```python
from pathlib import Path
```

```python
def input_pdf_setup(uploaded_file):
    if uploaded_file is not None:
        ## Convert the PDF to image
        path = Path("poppler-23.11.0\\Library\\bin")
        images= pdf2image.convert_from_path(uploaded_file,
                                            poppler_path= path)

        # first_page=images[0]
        pdf_parts = []

        for page_num, image in enumerate(images):
            # Convert to bytes
            img_byte_arr = io.BytesIO()
            image.save(img_byte_arr, format='JPEG')
            img_byte_arr = img_byte_arr.getvalue()

            pdf_part = {
                # "page_number": page_num + 1,  # Page numbers usually start from 1
                "mime_type": "image/jpeg",
                "data": base64.b64encode(img_byte_arr).decode()
            }
            data = [pdf_part]

            pdf_parts+=data

        return pdf_parts
    else:
        raise FileNotFoundError("No file uploaded")
```

```python
input_pdf_setup("2 pages.pdf")
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key= os.environ['GOOGLE_API_KEY'])

def get_gemini_response(prompt, pdf_content, input):
    model = genai.GenerativeModel('gemini-pro-vision')
    data = [PROMPT_RESUME_ABOUT ]+ pdf_content + ["This is the job description:"] + [input]
    response = model.generate_content(data)
    return response.text
```

```python
get_gemini_response(PROMPT_RESUME_ABOUT, input_pdf_setup("2 pages.pdf"), input)
```
